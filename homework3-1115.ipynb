```python
import pandas as pd
import numpy as np
from pandas import datetime
from matplotlib import pyplot as plt

"""
Load AirQualityUCI Data
"""

def parser(x):
    return datetime.strptime(x, '%Y-%m-%d %H:%M:%S')

input_file='./data/AirQualityUCI_refined.csv'

df=pd.read_csv(input_file,
              index_col=[0],
              parse_dates=[0],
              date_parser=parser)

df.head()
```

```python
df['PT08.S3(NOx)'].plot()
```

```python
df['PT08.S3(NOx)'].interpolate(inplace=True)
```

```python
pt = df['PT08.S3(NOx)'].copy()
pt.interpolate(inplace=True)
```

```python
plt.plot(df['PT08.S3(NOx)'],label='original',zorder=2)
plt.plot(pt, label='linear interploation', zorder=1)
plt.legend(loc='best')
```

```python
plt.boxplot(pt)
plt.title("Detecting outliers using Boxplot")
plt.xlabel('PT08.S3(NOx)')
```

```python
corr_matrix=df.corr()
corr_matrix
```

```python
rh = df['RH'].copy().interpolate()
```

```python
plt.scatter(pt, rh, s=12, c='black')
plt.xlabel('PT08.S3(NOx)')
plt.ylabel('RH')
plt.title("Detecting outliers using Boxplot")
```

```python
nmhc = df['PT08.S2(NMHC)'].copy().interpolate()
```

```python
plt.scatter(pt, nmhc, s=12, c='black')
plt.xlabel('PT08.S3(NOx)')
plt.ylabel("NMHC")
```

```python
q1 = pt.quantile(0.25)
median = pt.quantile(0.5)
q3 = pt.quantile(0.75)
print(q1, median, q3)
```

```python
iqr = q3-q1
upper_fence = q3 + 1.5*iqr
lower_fence = q1 - 1.5*iqr
print(upper_fence, lower_fence)
```

```python
outliers = pt.loc[(pt > upper_fence) | (pt < 0)]
print(outliers)
```

```python
mask = pt.index.isin(outliers.index)
```

```python
pt[~mask]
```

```python
plt.plot(pt[~mask], label='normal', color='blue',
    marker='o', markersize=3, linestyle='None')
plt.plot(outliers, label='outliers', color='red',
    marker='x', markersize=3, linestyle='None')
plt.legend(loc='best')
```

```python
pt_refined = pt.copy()
pt_refined[mask] = np.nan
```

```python
pt_refined.interpolate(inpalce = True)
plt.plot(pt_refined)
```
